```python
import pandas as pd
from split import create_splits
import config
import numpy as np
from torch.utils.data import DataLoader
from dataset import FrameDataset

from train import train_one_class
from eval import evaluate_class
import config
import json
```

```python

all_results = []

for class_name in config.CLASS_NAMES:
    print(f"\n============================")
    print(f"Processing class: {class_name}")
    
    try:
        train_one_class(class_name)
    except Exception as e:
        print(f"Training failed for {class_name}: {e}")
        continue

    # Evaluate model
    try:
        result = evaluate_class(class_name)
        if result:
            all_results.append(result)
    except Exception as e:
        print(f"Evaluation failed for {class_name}: {e}")

with open("eval_results.json", "w") as f:
    json.dump(all_results, f, indent=2)

print("\n DONE — all models trained & evaluated. Results in eval_results.json")
```
